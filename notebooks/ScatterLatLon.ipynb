```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
data_dir = "/mnt/windows/Users/Public/Documents/ECE278/project/cleanData/"
df = pd.read_csv(data_dir + "unzipped/2021/Clean_AIS_2021_01_01.csv")
```

```python
figure(figsize=(10, 8), dpi=500)
plt.scatter(df.LON,df.LAT,s=.1)
plt.annotate("Long Beach",(-118.1937,33.7701),color='red')
plt.annotate("San Diego",(-117.1611,32.7157),color='red')

plt.title("Routes taken on Jan 1, 2021 off the Coast of Southern California")
plt.xlabel("Longitude (Degrees)")
plt.ylabel("Latitude (Degrees)")
```
